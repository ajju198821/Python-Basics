# Built in Data types in python

### 1. Numbers type / Numeric type

#### Int Data type

```python
num1 = 10
num2 = 20
num3 = num1+num2
print(num3)
```

#### Float Data type

```python
num1 = 10.5
num2 = 3.55
num3 = num1 + num2
print(num3)
```

#### Complex Data type

```python
num1 = 2   +  5.5j
num2 = 2.5 +  4j

num3 = num1 + num2
print(num3)
```

#### How to determine the type of variable in Python?

Python provides a function named **type()** to determine which type a variable or a value belongs to.

```python
num1 = 30
num2 = 2.3
num3 = 2. + 3j

print(type(num1))
print(type(num2))
print(type(num3))
```
