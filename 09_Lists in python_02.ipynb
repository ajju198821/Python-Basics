# Python List Methods





## 1. Python List: append() method

The append() method is used to add an item to the end of a list.

`list.append(item)`

**Parameter:**
- item - Add an item (number, string, list, etc.) at the end of the list.

**Equivalent Command of append() method:**

`a[len(a):] = [x]`



#### Example1: Add an element to a list

```python
# colors list
colors = ['Red', 'Green', 'Black']
print("Original list:",colors)  
print("\nAdd 'White' to the list:")
colors.append('White')
print("\nNew list:",colors)
```

#### Example 2: Adding list to a list



```python
# colors list
colors = ['Red', 'Green', 'Black']
nums = [1, 2, 3]
print("\nOriginal lists:") 
print(colors) 
print(nums) 
print("\nAppend nums to colors:")
colors.append(nums)
print("New list:",colors)
```

## 2. Python List: extend() Method

The `extend()` method is used to extend a list by appending all the items from the iterable.

`list1.extend(iterable)`

**Parameter:**

Required. Type -> any iterable (list, set, tuple, etc.)



#### Example 1: List `extend()` Method



```python
# colors list
color1 = ['Red', 'Green', 'Black']
print("\nFirst list:") 
print(color1)  
color2 = ['White', 'Orange']
print("\nSecond list:") 
print(color2)
print("\nAdd the second list to the first list:")
color1.extend(color2)
print(color1)
```

#### Example 3: Add each element of a string to a list




```python
# colors list
colors = ['Red', 'Green', 'Black']
print("\nOriginal list:") 
print(colors)  
text = "Pink"
print("\nString to add:")
print(text)
print("\nAdd each element of the said string to the colors list:")
colors.extend(text)
print(colors)
```

## 3. Python List: clear() Method

The `clear()` method is used to remove all items from a given list.

`list.clear()`



#### Example1: clear() method



```python
# colors list
colors = ['Red', 'Green', 'Black']
print("\nOriginal list:") 
print(colors) 
print("\nClearing the said list:")
colors.clear()
print("New list:",colors)
```

### Example2: Clearing list using `del` statement

```python
# colors list
colors = ['Red', 'Green', 'Black']
print("\nOriginal list:") 
print(colors) 
print("\nClearing the said list using del statement:")
del colors[:]
print("New list:",colors)
```

## 4. Python List: copy() Method

The `copy()` method returns a copy of a given list.

`new_list = list.copy()`


#### Example 1: Copying a List



```python
# colors list
colors = ['Red', 'Green', 'Black']
print("\nOriginal list:") 
print(colors) 
print("\nCopying the said list:")
new_colors = colors.copy()
print("\nAdd an element to list:")
new_colors.append('White')
print("New list:",new_colors)
print("\nOriginal list:") 
print(colors) 
```

#### Example 2: Copy List using slicing syntax

```python
# colors list
colors = ['Red', 'Green', 'Black']
print("\nOriginal list:") 
print(colors) 
print("\nCopying the said list:")
new_colors = colors[:]
print("\nAdd an element to list:")
new_colors.append('White')
print("New list:",new_colors)
print("\nOriginal list:") 
print(colors) 
```

## 5. Python List: count() Method

The `count()` method is used to get the number of times a specified element appears in a given list.

`list.count(element)`

**Parameter:**

`element` - Required. The element is to be counted.. Type -> any type (string, number, list, tuple, etc.).

- Return value from `count()` method

The number of times an element appears in the list is returned.

### Example 1: Use of count() method



```python
# colors list
colors = ['Red', 'Green', "Red",'Black', 'Red', 'Green',"Red","Red"]
print("\nOriginal list:") 
print(colors)  
print("\nNumber of times the value 'Red' appears in the colors list:")
ctr = colors.count("Red")
print(ctr)
```

### Example 2: Count tuple and list elements inside list

```python
nums = [[1,2], [1,2,3], [1,2],['1','2'],[3,4,5], [1,2]] 
print("\nOriginal list:")
print(nums)
print("\nNumber of times the list '[1,2]' appears in the said list:")
ctr = nums.count([1,2])
print(ctr)
```

## 6.Python List: index() Method

The `index()` method is used to get the zero-based index in the list of the first item whose value is given.

`list.index(element, start, end)`

**Parameters:**

- **element -** The element to be searched.
- **start (optional), end (optional) -** The optional arguments `start` and `end` are interpreted as in the slice notation and are used to limit the search to a particular subsequence of the list.

**Return Value from List index()**

- Return zero-based index in the list of the first item
- Raises a Value Error if there is no such item.

### Example 1: Find the index of an element in a list



```python
# colors list
colors = ['Red', 'Green', 'Black', 'Green']
print("\nOriginal list:") 
print(colors)  
print("\nPosition at the first occurrence of the specified value:")
p = colors.index("Green")
print(p)
```

### Example 2: Find index of an element after a given position in a list



```python
# fruits list
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print("\nOriginal list:") 
print(fruits)  
print("\nPosition of 'apple' after the specified position:")
p = fruits.index("apple", 4)
print(p)
```

### Example 3: Find index of an element within a certain range in a list



```python
# fruits list
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
print("\nOriginal list:") 
print(fruits)  
print("\nPosition of 'apple' within a certain range:")
p = fruits.index("apple", 2, 6)
print(p)

# print("\nPosition of 'apple' within a certain range:")
# p = fruits.index("apple", 2, 3)
# print(p)
# Will raise ValueError: 'apple' is not in list
```

## 7.Python List: insert() Method

The `insert()` method is used to insert an item at a given position in a list.

`list.insert(i, element)`

**Parameter:**

- index - The first argument is the index of the element before which to insert.
- element - Element to be inserted in the list.

### Example 1: Insert a given value at the specified position



```python
# colors list
colors = ['Red', 'Green', 'Black']
print("\nOriginal list:") 
print(colors)  
print("\nInsert 'White' at the second position in the said list:")
colors.insert(1, "White")
print(colors)
```

### Example 2: Insert a list to the list of lists in the specified position

```python
# colors list
nums = [[1,2,3], [1,2], [4,5,6], [4,5]]
print("\nOriginal list:") 
print(nums) 
n = [-1,0,2,'Aziz',['Abdul']]
nums.insert(2, n)
print("\nInsert a list at the second position of the said list of lists:")
print(nums)
```

## 8. Python List: pop() Method

The `pop()` method is used to remove the item at the given position in a list, and return it. If no index is specified, `a.pop()` removes and returns the last item in the list.

`list.pop([i])`


### Example 1: Remove an item at the given index from the list


```python
colors = ['Red', 'Green', 'Orange', 'Pink']
print("Original List:")
print(colors)
print("\nRemove and return the 3rd item:")
return_item = colors.pop(2)
print('\nReturn Value:', return_item)
print('\nUpdated List:', colors)
```

The  `remove()` method is used to remove the first occurrence of a given element from a list.

## 10. Python List: reverse() Method

The `reverse()` method is used to reverse the elements of a given list.

`list.reverse()`

### Example 1: Reverse a List



```python
colors = ['Red', 'Green', 'Orange', 'Pink']
print("\nOriginal List:")
print(colors)
colors.reverse()
print('\nUpdated List:', colors)
```

## 11. Python List: sort() Method

The `sort()` method is used to sort the items of a given list.

### Example 1: Sort a list in ascending/descending order


```python
colors = ['Red', 'Green', 'Orange', 'Pink']
print("Original List:")
print(colors)
colors.sort()
print('\nSort the said list in ascending order:')
print(colors)
print('\nSort the said list in descending order:')
colors.sort(reverse=True)
print(colors)
```
